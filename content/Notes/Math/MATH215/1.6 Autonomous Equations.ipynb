# Autonomous Equations

When dealing with problems of the form:
$$
\frac{dx}{dt} = f(x)
$$
they are considered autonomous as they are independent of the variable being differentiated.

Because of this, these equations can have equilibrium solutions where for as long as $x$ is at a critical point such that $f(x) =0$ then the equation is in equilibrium. Namely, each critical point corresponds to an equilibrium solution.

Furthermore, these equilibria are also called ==stable== if for some critical point $x =a$ we have that:

$$
\lim_{ t \to \infty } x(t) =a
$$
Now, consider the *logistic equation* used for modelling populations where $M$ is the maximum sustainable population.

$$
\frac{dx}{dt} = kx(M-x)
$$
Graphing the slope field for a similar equation which is autonomous we see:

$$
x' = 0.1x(5-x)
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# ODE: x' = 0.1x(5 - x)
def f(t, x):
    return 0.1 * x * (5 - x)

# Grid for slope field
t_vals = np.linspace(0, 20, 25)
x_vals = np.linspace(-1, 7, 25)
T, X = np.meshgrid(t_vals, x_vals)

# Direction vectors
dT = np.ones_like(T)
dX = f(T, X)

# Normalize so arrows have similar length
M = np.sqrt(dT**2 + dX**2)
dT = dT / M
dX = dX / M

plt.figure(figsize=(10, 6))
plt.quiver(T, X, dT, dX, angles="xy")

# Equilibrium solutions
plt.axhline(0, linestyle="--", linewidth=1, label="x = 0")
plt.axhline(5, linestyle="--", linewidth=1, label="x = 5")

plt.title(r"Slope field for $x' = 0.1x(5-x)$")
plt.xlabel("t")
plt.ylabel("x")
plt.xlim(0, 20)
plt.ylim(-1, 7)
plt.grid(True)
plt.legend()
plt.show()
```

From the slope field we can see that:
$$
\lim_{ t \to \infty } x(t) = \begin{cases}
5 \quad x(0)>0 \\
0 \quad x(0) = 0  \\
DNE \quad x(0) < 0
\end{cases}
$$

In brief, we can observe that $x=5$ is clearly a stable point as $x = 5$ is a critical point and the slope field converges to it, whereas $x=0$ is unstable as the slope field diagram diverges from it.
